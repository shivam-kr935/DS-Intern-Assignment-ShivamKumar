```python
# Imputing missing values using Simple Imputer

from sklearn.impute import SimpleImputer

# Select only numeric columns with missing values
missing_cols = df.columns[df.isnull().any()].tolist()
numeric_cols = df[missing_cols].select_dtypes(include=['float64', 'int64']).columns

# Apply median imputation
imputer = SimpleImputer(strategy='median')
df[numeric_cols] = imputer.fit_transform(df[numeric_cols])

print("Missing values imputed using median strategy")
```

```python
#Missing values are filled using median imputation technique
print(df[numeric_cols].isnull().sum().sum())
```

#No missing values found and now it is cleaned data

```python
#Outlier detection
Q1 = df['equipment_energy_consumption'].quantile(0.25)
Q3 = df['equipment_energy_consumption'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove or cap outliers
df_cleaned = df[(df['equipment_energy_consumption'] >= lower_bound) &
                        (df['equipment_energy_consumption'] <= upper_bound)]

print(f"After Removing Outliers: {df.shape}")
```
