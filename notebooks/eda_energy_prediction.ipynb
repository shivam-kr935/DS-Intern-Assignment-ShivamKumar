```python
#Importing basic library
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Loading the dataset
df = pd.read_csv('../data/data.csv')
df.head()
```

```python
#For viewing all features let me set pandas max view option
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
df.head()
```

```python
#Checking basic info
df.shape
```

```python
#Checking missing values
df.isnull().sum()
```

```python
#Percentage of missing values
(df.isnull().sum() / len(df)) * 100
```

```python
#basic stats
df.describe()
```

```python
#checking target variable distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['equipment_energy_consumption'], kde=True, bins=30)
plt.title('Distribution of Equipment Energy Consumption')
plt.xlabel('Energy')
plt.ylabel('Frequency')
plt.show()
```

```python
#Checking correaltion between features and target variables
corr=df.corr(numeric_only=True)
print(corr)
```

```python
#correaltion heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
#Converting the feature timestamp into features
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.dayofweek
df['month'] = df['timestamp'].dt.month
```
